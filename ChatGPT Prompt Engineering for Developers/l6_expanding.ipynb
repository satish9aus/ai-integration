In this lesson, you will generate customer service emails that are tailored to each customer's review.

## Transcript summary

**What is expanding?** Taking a short input (instructions, a review, a list of topics) and generating a longer output (email, essay, reply). Great for brainstorming — but use responsibly; the same capability can generate spam at scale.

**Sentiment-aware email generation**:
- Pass both the review text and its pre-extracted sentiment into the prompt
- Instruct the model to branch on sentiment: thank for positive/neutral, apologize + suggest support for negative
- Always sign AI-generated customer-facing text as an AI agent — transparency is essential

**Temperature — the key new concept**:

| Temperature | Behaviour | Use when |
|---|---|---|
| `0` | Always picks the most likely next token → deterministic, repeatable | Production systems, reliable pipelines |
| `0.7` | Samples from less likely tokens → varied output each run | Creative tasks, brainstorming, draft generation |
| High (→1) | Even more random, "distractible" | Maximum variety, exploratory use |

**Rule of thumb**: build and test at `temperature=0`; raise it only when variety is a feature, not a bug.
