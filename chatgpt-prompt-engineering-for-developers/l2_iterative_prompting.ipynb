# L2: Iterative Prompting

This notebook shows how to improve model outputs through an iterative process: write a prompt, inspect the result, then refine the prompt and response.

## Why prompt iteration matters

- When building apps with LLMs, the first prompt is rarely the final prompt.
- That is okay — what matters is the process for improving the prompt.
- A good prompt process is:
  1. decide what task you want to solve,
  2. write a first prompt,
  3. run it and inspect the result,
  4. refine the prompt,
  5. repeat until it works well.
- This is similar to machine learning development: try something, inspect the result, fix it, and try again.
- There is not one perfect prompt for every situation.
- The important part is having a process to develop a prompt that works for your application.

In this notebook example, we will:
- start with a first prompt,
- judge the output,
- make the prompt shorter or more precise,
- and continue refining until the answer is better.

```python
from helper import get_completion

# Example helper call ready for iterative refinement
```

## Step 1: Start with an initial prompt

Write a first prompt and inspect the model's response.

```python
initial_prompt = '''
Write a short product description for a smartwatch that is simple, friendly, and useful for a busy professional.
'''

initial_response = get_completion(initial_prompt)
print(initial_response)
```

## Step 2: Evaluate the output and ask for improvement

Use the model to critique the first answer and suggest changes.

```python
feedback_prompt = f'''
The following product description is intended for a busy professional.
Please do the following:
1. Identify what is good about the description.
2. Suggest two improvements to make it more concise, engaging, and action-oriented.
3. Rewrite the description using those improvements.

Original description:
"""
{initial_response}
"""
'''

feedback = get_completion(feedback_prompt)
print(feedback)
```

## Step 3: Refine the prompt and run again

Update the prompt with clearer instructions and run it again.

```python
refined_prompt = '''
Write a concise, engaging product description for a smartwatch that is designed for a busy professional.
Be action-oriented, highlight the main benefits, and keep the tone friendly and polished.
Use one paragraph with 2-3 sentences.
'''

refined_response = get_completion(refined_prompt)
print(refined_response)
```

## Step 4: Compare the original and refined outputs

Compare the two responses and choose the better one.

```python
compare_prompt = f'''
Compare the two smartwatch descriptions below.
1. Identify which description is better overall for a busy professional.
2. Explain why.
3. If possible, provide a final polished version that combines the best elements of both.

Description A:
"""
{initial_response}
"""

Description B:
"""
{refined_response}
"""
'''

comparison = get_completion(compare_prompt)
print(comparison)
```
