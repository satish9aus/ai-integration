# L1: Language Models, the Chat Format and Tokens

## Study Notes

**How LLMs are trained**:
1. **Supervised learning** on hundreds of billions of words — model learns to predict the next word given a sentence prefix (takes months on a supercomputer)
2. **Fine-tuning** on a smaller set of instruction→response examples (days, modest compute)
3. **RLHF** (Reinforcement Learning from Human Feedback) — humans rate outputs on helpfulness/honesty/harmlessness; model is tuned to favour higher-rated outputs

**Base LLM vs. Instruction-Tuned LLM**:

| | Base LLM | Instruction-Tuned LLM |
|---|---|---|
| Prompt: *What is the capital of France?* | May list more quiz questions | *The capital of France is Paris.* |
| Behaviour | Completes text statistically | Follows instructions |

**Tokens, not words**: LLMs predict the next *token*, not the next word.
- Common words → 1 token; rare/long words → multiple tokens (e.g. `lollipop` → `l` + `oll` + `ipop`)
- ~1 token ≈ 4 characters ≈ ¾ of a word (English)
- GPT-3.5-turbo limit: ~4,000 tokens (input + output combined)
- **Trick**: insert dashes between letters (`l-o-l-l-i-p-o-p`) to force single-character tokens → model can reverse them correctly

**Chat format** — three roles:
- `system` → sets overall tone/persona (invisible to end user)
- `user` → the human's request
- `assistant` → the model's response (can be injected to continue a conversation)

**Token counting**: The API response includes `usage.prompt_tokens`, `usage.completion_tokens`, and `usage.total_tokens` — useful for staying within context limits.

**Secure API key handling**: Use `dotenv` + a `.env` file rather than hardcoding the key in the notebook.

**Prompting vs. traditional ML** (unstructured/text tasks):

| | Traditional supervised ML | Prompting |
|---|---|---|
| Labelled data | Hundreds of examples, weeks to collect | None needed |
| Model training | Days–months | Minutes–hours |
| Deployment | Weeks | API call |
| Total time to value | Months | Hours–days |

> Caveat: prompting works best for unstructured data (text, vision). Tabular/numerical data still benefits more from traditional ML.

## Helper function (chat format)

````python
messages = [
    {'role': 'system',
     'content': """You are an assistant who\
 responds in the style of Dr Seuss."""},
    {'role': 'user',
     'content': """write me a very short poem\
 about a happy carrot"""},
]
response = get_completion_from_messages(messages, temperature=1)
print(response)
````

````python
# length
messages = [
    {'role': 'system',
     'content': 'All your responses must be \
one sentence long.'},
    {'role': 'user',
     'content': 'write me a story about a happy carrot'},
]
response = get_completion_from_messages(messages, temperature=1)
print(response)
````

````python
# combined
messages = [
    {'role': 'system',
     'content': """You are an assistant who \
responds in the style of Dr Seuss. \
All your responses must be one sentence long."""},
    {'role': 'user',
     'content': """write me a story about a happy carrot"""},
]
response = get_completion_from_messages(messages, temperature=1)
print(response)
````

````python
messages = [
    {'role': 'system',
     'content': """You are an assistant who responds\
 in the style of Dr Seuss."""},
    {'role': 'user',
     'content': """write me a very short poem \ 
 about a happy carrot"""},
]
response, token_dict = get_completion_and_token_count(messages)
print(response)
print(token_dict)
````

## A note about the backslash
In the course, we are using a backslash `\` to make the text fit on the screen without inserting newline `'\n'` characters.
GPT-3 isn't really affected whether you insert newline characters or not. But when working with LLMs in general, you may consider whether newline characters in your prompt may affect the model's performance.
